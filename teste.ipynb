```python
import MetaTrader5 as mt5

# Parâmetros de login da conta
LOGIN = 12345678               # Substitua pelo número da sua conta
SERVER = "NomeServidor"       # Ex: "MetaQuotes-Demo"
PASSWORD = "sua_senha_aqui"   # Sua senha da conta MT5

# Inicializa o MT5 com login manual
if not mt5.initialize(login=LOGIN, server=SERVER, password=PASSWORD):
    print(f"❌ Erro ao iniciar MetaTrader 5: {mt5.last_error()}")
    quit()

# Verifica se a conta está conectada
account_info = mt5.account_info()
if account_info is None:
    print("⚠️ Não foi possível obter informações da conta. Verifique as credenciais ou o terminal MT5.")
    mt5.shutdown()
    quit()

# Exibe informações básicas da conta
print("✅ Conexão estabelecida com sucesso!")
print(f"Login: {account_info.login}")
print(f"Servidor: {account_info.server}")
print(f"Nome: {account_info.name}")
print(f"Saldo: {account_info.balance}")
print(f"Equity: {account_info.equity}")
print(f"Margem Livre: {account_info.margin_free}")
print(f"Nível de Margem: {account_info.margin_level}")

# Encerra a conexão
mt5.shutdown()

```
